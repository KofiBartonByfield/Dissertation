# Combine Crime Data

### Load libraries 

```python

import os
import pandas as pd
import matplotlib.pyplot as plt

path = 'data/police_zips/unzipped_data'    

```

## Extract the names of the CSV's

```python
csv_names = []
for folder in os.listdir(path):
        
            folder_contents = os.listdir(f'{path}/{folder}')
            
            for file in folder_contents:
                
                if file.endswith('street.csv'):
                    
                    csv_names.append(f'{path}/{folder}/{file}')
        
        

csv_names
```

## 

```python
dataframes = []

for csv in csv_names:

    new_df = pd.read_csv(csv)
    
    dataframes.append(new_df)


df = pd.concat(dataframes, ignore_index=True)


display(df)
```

### Save to new df to CSV

```python
df.to_csv('data/crime.csv')
```

```python
df['Month'] = pd.to_datetime(df['Month'], format = "%Y-%m")

display(df)
```

```python
crime_counts = df['Month'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(crime_counts.index, crime_counts.values, linestyle='-')

# Labels and title
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Crime Count per Month')
plt.xticks(rotation=45)
plt.grid()

# Show plot
plt.show()
```
