```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/london_street_crime_sep14_nov24.csv')

display(df)
```

```python
df['Date'] = pd.to_datetime(df['Month'], format = '%Y-%m')

df.set_index('Date', inplace=True)

display(df)
```

```python

lockdown_periods = [
    ('2023-05-02', '2023-05-02'),  #public order ACT 23
]


# Resample data to weekly counts
daily_counts = df.resample('ME').size()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(daily_counts, label="Weekly Stops and Searches")

# Highlight lockdown periods
for start, end in lockdown_periods:
    plt.axvspan(pd.to_datetime(start), pd.to_datetime(end), alpha=0.3, label="Public order act", color='red')

# Add labels, title, and legend
plt.title("Occurrences of Crimes Over Time ")
plt.xlabel("Date")
plt.ylabel("Number of Crimes")
plt.legend(loc='upper left')
# plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

plt.figure(figsize=(12, 6))
           
# Resample data to weekly counts
for dept in df['Crime type'].value_counts().index[:2]:
    daily_counts = df[df['Crime type'] == dept].resample('ME').size()

    # Plot the data
    plt.plot(daily_counts, label=dept)

plt.axvspan(pd.to_datetime('2023-05-02'), pd.to_datetime('2023-05-02'),  label="Public order act", color='red')
plt.axvspan(pd.to_datetime('2022-10-01'), pd.to_datetime('2022-10-01'), label="JSO", color='green')
plt.axvspan(pd.to_datetime('2020-5-28'), pd.to_datetime('2020-5-28'), label="BLM", color='black')
plt.axvspan(pd.to_datetime('2022-3-27'), pd.to_datetime('2022-3-27'),  label="Ukraine", color='Blue')
plt.axvspan(pd.to_datetime('2024-07-30'), pd.to_datetime('2024-08-05'), label="Far right", color='pink')


plt.axvspan(pd.to_datetime('2020-03-23'), pd.to_datetime('2020-06-01'), alpha=0.3, label="Covid-19", color='red')




# Add labels, title, and legend
plt.title("London Crime rates by crime type")
plt.xlabel("Date")
plt.ylabel("Number of Stops/Searches")
plt.legend(loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
