```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/stop_and_search_cleaned.csv')

display(df)
```

```python
df['Date'] = df['Date'].str.slice(0, 16)
df['Date'] = pd.to_datetime(df['Date'], format = "%Y-%m-%d %H:%M")

df.set_index('Date', inplace=True)

display(df)
```

# Visualise the Data


## National Lockdowns

```python

# Assume lockdown periods in the UK (adjust dates as necessary)
lockdown_periods = [
    ('2020-03-23', '2020-06-01'),  # First lockdown
    ('2020-09-05', '2020-11-02'),  # Second lockdown
    ('2021-01-06', '2021-07-01')   # Third lockdown
]




# Resample data to weekly counts
daily_counts = df.resample('ME').size()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(daily_counts, label="Weekly Stops and Searches")

# Highlight lockdown periods
for start, end in lockdown_periods:
    plt.axvspan(pd.to_datetime(start), pd.to_datetime(end), alpha=0.3, label="COVID", color='red')

# Add labels, title, and legend
plt.title("Occurrences of Stops and Searches Over Time ")
plt.xlabel("Date")
plt.ylabel("Number of Stops/Searches")
plt.legend(loc='upper left')
# plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

## By dept.

```python

plt.figure(figsize=(12, 6))
           
# Resample data to weekly counts
for dept in df.Police_Dept.value_counts().index[:5]:
    daily_counts = df[df['Police_Dept'] == dept].resample('ME').size()

    # Plot the data
    plt.plot(daily_counts, label=dept)


# Add labels, title, and legend
plt.title("Stops and Searches Over Time by Police Dept. ")
plt.xlabel("Date")
plt.ylabel("Number of Stops/Searches")
plt.legend(loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```

# London

```python
london_df = df[df['Police_Dept'] == 'metropolitan']
london_df
```

```python
daily_counts = london_df.resample('ME').size()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(daily_counts, label="Monthly Stops and Searches")

plt.axvspan(pd.to_datetime('2023-05-02'), pd.to_datetime('2023-05-02'),  label="Public order act", color='red')
plt.axvspan(pd.to_datetime('2022-10-01'), pd.to_datetime('2022-10-01'), label="JSO", color='green')
plt.axvspan(pd.to_datetime('2020-5-28'), pd.to_datetime('2020-5-28'), label="BLM", color='black')
plt.axvspan(pd.to_datetime('2022-3-27'), pd.to_datetime('2022-3-27'),  label="Ukraine", color='Blue')
plt.axvspan(pd.to_datetime('2024-07-30'), pd.to_datetime('2024-08-05'), label="Far right", color='pink')


plt.axvspan(pd.to_datetime('2020-03-23'), pd.to_datetime('2020-06-01'), alpha=0.3, label="Covid-19", color='red')





plt.legend()
plt.title('London Stops & Searches')
plt.show()
```

```python
ethnicities = london_df['Self_defined_ethnicity'].dropna().unique()

# Set up the plot
plt.figure(figsize=(12, 6))
ethnicity_to_skip = ethnicities[2]  # Index 2 corresponds to the third ethnicity


# Plot each self-defined ethnicity group, skipping the third ethnicity
for ethnicity in ethnicities[:4]:
    if ethnicity == ethnicity_to_skip:
        continue  # Skip the 3rd ethnicity
    ethnicity_data = london_df[london_df['Self_defined_ethnicity'] == ethnicity].resample('ME').size()
    plt.plot(ethnicity_data, label=f"{ethnicity} Stops and Searches")

# Highlight key events on the plot
plt.axvspan(pd.to_datetime('2023-05-02'), pd.to_datetime('2023-05-02'), label="Public order act", color='red')
plt.axvspan(pd.to_datetime('2022-10-01'), pd.to_datetime('2022-10-01'), label="JSO", color='green')
plt.axvspan(pd.to_datetime('2020-5-28'), pd.to_datetime('2020-5-28'), label="BLM", color='black')
plt.axvspan(pd.to_datetime('2022-3-27'), pd.to_datetime('2022-3-27'), label="Ukraine", color='blue')
plt.axvspan(pd.to_datetime('2024-07-30'), pd.to_datetime('2024-08-05'), label="Far right", color='pink')
plt.axvspan(pd.to_datetime('2020-03-23'), pd.to_datetime('2020-06-01'), alpha=0.3, label="Covid-19", color='grey')

# Customise plot
plt.legend()
plt.title('London Stops & Searches by Self-Defined Ethnicity')
plt.xlabel('Date')
plt.ylabel('Count of Stops and Searches')

# Show the plot
plt.show()
```
