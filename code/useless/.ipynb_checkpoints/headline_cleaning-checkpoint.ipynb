```python
import pandas as pd
import os

```

```python

# List of file paths for the articles
file_paths = os.listdir('../data/headlines')
file_paths
```

```python
data = pd.read_csv(f"../data/headlines/{file_paths[4]}")
headlines = data['title'].dropna().tolist()  # Ensure no NaN values
```

```python
headlines[:10]
```

```python
def tokenize(text):
    return text.lower().split()

tokens = [tokenize(headline) for headline in headlines]

tokens
```

```python
positive_words = {'good', 'great', 'excellent', 'positive', 'happy', 'success', 'win'}
negative_words = {'murder', 'after', 'charged', 'death', 'arrested', 'police', 'probe', 'stabbed'}

def get_sentiment_score(words):
    score = 0
    for word in words:
        if word in positive_words:
            score += 1
        elif word in negative_words:
            score -= 1
    return score

# Analyse sentiment for each headline
sentiment_scores = [get_sentiment_score(token) for token in tokens]
sentiment_scores
```

```python
def classify_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

sentiment_categories = [classify_sentiment(score) for score in sentiment_scores]
sentiment_categories
```

```python

data = list(zip(headlines, sentiment_categories))
df = pd.DataFrame(data, columns=["headlines", "sentiment_categories"])


negative_headlines = df[df['sentiment_categories'] == 'Negative'].headlines
```

```python
negative_headlines
```

```python
from collections import Counter

combined_text = " ".join(negative_headlines)

words = combined_text.lower().split()


top_10 = dict(Counter(words).most_common(20))

list(top_10)


# need a list of top 10 that are not in this list:negative_words
# then iteratively build up a list of negative words that are not connectors such as of, in , to etc
```

```python
from collections import Counter

# Combine all negative headlines into one string
combined_text = " ".join(negative_headlines)

# Tokenize the text (convert to lowercase, split by spaces)
words = combined_text.lower().split()

# Get the top 20 most common words
top_10 = dict(Counter(words).most_common(20))

# Define a list of common words to exclude (connectors)
connectors = ['of', 'in', 'to', 'the', 'and', 'for', 'on', 'a', 'an', 'at', 'with', 'by', 'as', 'that', 'it', 'this', 'man']

# Filter out the words from top_10 that are in connectors list
top_10_filtered = {word: count for word, count in top_10.items() if word not in connectors}

# Now `top_10_filtered` contains top 10 words excluding common connectors
print("Top 10 words (excluding connectors):", top_10_filtered)

# Iterate to build up the list of negative words, excluding connectors
negative_words = []  # List of negative words (start with an empty list)

for word in top_10_filtered:
    if word not in connectors:  # Ensure it's not a connector
        negative_words.append(word)

print("Negative words list:", negative_words)

```
