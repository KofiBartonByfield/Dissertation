# Match Self Defined Ethnicities

```python
# mapping
category_mapping = {
    "White - English/Welsh/Scottish/Northern Irish/British": "White/British",
    "White - White British (W1)": "White/British",
    "White - Any other White background": "Other White",
    "White - Any other White ethnic background (W9)": "Other White",
    "Black/African/Caribbean/Black British - Any other Black/African/Caribbean background": "Black/Black British - Other",
    "Black/African/Caribbean/Black British - African": "Black/Black British - African",
    "Black/African/Caribbean/Black British - Caribbean": "Black/Black British - Caribbean",
    "Asian/Asian British - Any other Asian background": "Asian/Asian British - other",
    "Black or Black British - Any other Black ethnic background (B9)" : "Black/Black British - Other",
    "Black or Black British - African (B2)" : "Black/Black British - African",
    "Black or Black British - Caribbean (B1)": "Black/Black British - Caribbean",
    "Asian or Asian British - Any other Asian ethnic background (A9)": "Asian/Asian British - other",
    "Asian or Asian British - Pakistani (A2)": "Asian/Asian British - Pakistani",
    "Asian or Asian British - Bangladeshi (A3)": "Asian/Asian British - Bangladeshi",
    "Mixed - White and Black Caribbean (M1)": "Mixed/Multiple ethnic groups - White and Black Caribbean",
    "Mixed - White and Black African (M2)": "Mixed/Multiple ethnic groups - White and Black African",
    "Mixed - White and Asian (M3)": "Mixed/Multiple ethnic groups - White and Asian",
    "Asian or Asian British - Indian (A1)": "Asian/Asian British - Indian",
    "Mixed - Any other Mixed ethnic background (M9)": "Mixed/Multiple ethnic groups - Any other Mixed/Multiple ethnic background",
    "Not Stated (NS)": "Not Stated",
    "Other ethnic group - Not stated": "Not Stated",
    "Chinese or other ethnic group - Chinese (O1)":"Chinese/Other ethnic group",
    "Chinese or other ethnic group - Any other ethnic group (O9)":"Chinese - other",
    "White - White Irish (W2)":"White - Irish",

}

# apply the mapping
df['Self_defined_ethnicity'] = df['Self-defined ethnicity'].replace(category_mapping)
df['Self_defined_ethnicity'].value_counts()
```
